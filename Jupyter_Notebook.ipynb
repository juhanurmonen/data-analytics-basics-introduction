# Jupyter Notebook

**Jupyter Notebook** is a practical tool for writing and running Python programs. Jupyter Notebook is automatically installed in pursuance of installation of Anaconda. 

### Open Jupyter Notebook

Open Jupyter Notebook from Start menu under Anaconda. It is also possible to create an icon on desktop. 

The front page of Jupyter Notebook starts on a new tab called **Home** of your default browser. 

### Create a new notebook

You can create a new empty notebook from top right corner operation of the front page called **New Python 3**. The new notebook opens on a new tab.

At the top of the new notebook you find several tool buttons and menus worth knowing. The notebook can be renamed by clicking the old or default name at the top row of the notebook. Renaming can also be done by choosing **Rename** operation from the **File** menu.

### Write code

Notebooks consist of cells. Into a cell you can write Python code and comments. Comment rows always start with the character $\sharp$. Code cells can be run in different ways

- Using the operations in the **Cell** menu at the top of your notebook.
- Using button on the tool bar.
- Using keyboard shortcuts **ctrl-enter** or **shift-enter**. The latter moves the cursor to the next cell. More keyboard shortcuts can be found by performing internet search **Jupyter keyboard shortcuts**. Try yourself: <a href="https://www.google.com/search?q=Jupyter+keyboard+shortcuts">Google search</a> &ndash; <a href="https://www.qwant.com/?q=Jupyter+keyboard+shortcuts">Qwant search</a> &ndash; <a href="https://www.bing.com/search?q=Jupyter+keyboard+shortcuts">Bing search</a> &ndash; <a href="https://www.duckduckgo.com/?q=Jupyter+keyboard+shortcuts">DuckDuckGo search</a>.

Possible outputs, warnings and error messages appear below the cell that was ran. You can any time change the code in cell and run the code again.

New cells can be added from the **Insert** menu and cells can be deleted from the **Edit** menu. 

Variables and other information saved when code was run earlier is available for the whole session. Therefore you can run part of the program in one cell and continue in another cell.

In case you modify and run cells that appear earlier in your notebook, unexpected errors may occur. When necessary, information saved during the session can be purged by choosing the **Kernel** menu and there either the operation **Restart & Clear Output** or the operation **Restart & Run All**. Each notebook has its own **Kernel**. The operation **Restart** has effect only on the active notebook.

When you open a previous started notebook and continue working on it, run previous written code by selecting **Restart & Run All** from the **Kernel** menu.

### Save the notebook

Notebook is periodically saved automatically. You can also push Jupyter to save the notebook using operations in the **File** menu. There is an icon for saving the notebook on top of the page. Also keyboard shortcut **ctrl-s** is good to remember. To be on the safe side, it is good to save the notebook before closing it. This way the latest version of the notebook is saved.

Notebooks are saved on Jupyter's default directory. On computers running a Windows operating system, the default directory is usually of the form C://Users/userid/. Jupyter Notebook files are recognised from file extension **.ipynb**. Saved notebooks are found on the Jupyter front page (**Home** tab) and a notebook can be opened on its own tab by clicking the notebook's name.

New subdirectories can be created under the default folder. It is posiible to move to another directory by clicking the (sub)directory's name on the Jupyter front page (**Home** tab). A new notebook is always saved on the directory which is selected when the notebook is created.
